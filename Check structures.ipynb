```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import igraph as ig
import math
from os import listdir
```

```python
componentsSet = set()

def getFile(folder, fileName):
    data = json.load(open(folder+"/"+fileName))
    dimensions = data['bounds']
    children = data['children']
    return dimensions, children

def listFiles(folder):
    files = listdir(folder)
    return files

def processComponentBlob(children):
    for child in children:
        clickable = child["clickable"]
        cClass = child["class"]
        label = child["componentLabel"]
        if("children" in child):
            processComponentBlob(child["children"])
        if(clickable == False):
            component = label+" "+cClass+" "+str(clickable)
            componentsSet.add(component)
    

def fetchUniqueComponents():
    folder = "hierarchies"
    files = listFiles("hierarchies")
    for file in files:
        bounds, children = getFile(folder, file)
        processComponentBlob(children)
```

```python
fetchUniqueComponents()
for c in componentsSet:
    print(c)
```

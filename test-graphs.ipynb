```python
import igraph as ig
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = ig.Graph()
print(g)
```

```python
g.add_vertices(3)
print(g)
```

```python
g.add_edges([(0,1), (1,2)])
print(g)
ig.summary(g)
```

```python
g.es['type'] = ["observe", "tap"]
g.vs['node_type'] = ["start", "button", "end"]
print(g)
```

```python
ig.plot(g, bbox=(0, 0, 100, 100))
```

```python
g.vs["label"] = g.vs["node_type"]
g.es["label"] = g.es["type"]
layout = g.layout("kk")
ig.plot(g, layout=layout, bbox=(0, 0, 200, 200), margin=20)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
ig.plot(g, target=axs[0])
axs[1].bar(x=[0, 1, 2], height=[1, 5, 3], color='tomato')
```

```python
g.get_adjacency()
```

```python
matr = [[0, 0, 0], [1, 0, 0], [0, 1, 0]]
gg = ig.Graph.Adjacency(matr, mode="directed")
```

```python
layout = gg.layout("kk")
ig.plot(gg, layout=layout, bbox=(0, 0, 200, 200), margin=20)
```

```python
layout = gg.layout("rt")
ig.plot(gg, layout=layout, bbox=(0, 0, 200, 200), margin=20)
```
